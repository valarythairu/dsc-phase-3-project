```python
#checking for outliers 
def outliers(df):
    # Dictionary for placing no. of outliers and the column name.
    outliers_dict = {}
    standardized_dict = {}

    # For loop for checking outliers in every column in the dataframe.
    for column in df.select_dtypes(include=[np.number]).columns:
        # Lower quantile
        Q1 = df[column].quantile(0.25)

        # Upper Quantile
        Q3 = df[column].quantile(0.75)

        # Finding IQR
        IQR = Q3 - Q1

        # Defining lower and upper bound
        lower_bound = Q1 - (1.5 * IQR)
        upper_bound = Q3 + (1.5 * IQR)
        
        # Finding outliers
        outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers_dict[column] = 'The number of outliers is', len(outliers)

        # Data without outliers
        standardized_numvotes = df[column][(df[column] >= lower_bound) & (df[column] <= upper_bound)]

        # Putting standardized values in a dict
        standardized_dict[column] = standardized_numvotes
        
    standardized_df = pd.DataFrame(standardized_dict)
    # Results
    return standardized_df
```

Feature Engineering 

This is done by creating new features from the features we have already in an attempt to understand the dataset better.

```python
# BMI Category
# Define function to get BMI category
def get_bmi_category(bmi):
    if bmi <= 18.5:
        return 'Underweight'
    elif bmi <= 24.9:
        return 'Normal weight'
    elif bmi <= 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Apply the function to create BMI_Category column
df_new['BMI_Category'] = df_new['BMI'].apply(get_bmi_category)

# Health Checkup Frequency
checkup_mapping = {'Within the past year': 4, 'Within the past 2 years': 2, 'Within the past 5 years': 1, '5 or more years ago': 0.2, 'Never': 0}
df_new['Checkup_Frequency'] = df_new['Checkup'].replace(checkup_mapping)

# Lifestyle Score
exercise_mapping = {'Yes': 1, 'No': 0}
smoking_mapping = {'Yes': 1, 'No': 0}
df_new['Lifestyle_Score'] = df_new['Exercise'].replace(exercise_mapping) - df_new['Smoking_History'].replace(smoking_mapping) + df_new['Fruit_Consumption']/10 + df_new['Green_Vegetables_Consumption']/10 - df_new['Alcohol_Consumption']/10

# Healthy Diet Score
df_new['Healthy_Diet_Score'] = df_new['Fruit_Consumption']/10 + df_new['Green_Vegetables_Consumption']/10 - df_new['FriedPotato_Consumption']/10

# Interaction Terms
df_new['Smoking_Alcohol'] = df_new['Smoking_History'].replace(smoking_mapping) * df_new['Alcohol_Consumption']
df_new['Checkup_Exercise'] = df_new['Checkup_Frequency'] * df_new['Exercise'].replace(exercise_mapping)

# Ratio of Height to Weight
df_new['Height_to_Weight'] = df_new['Height_(cm)'] / df_new['Weight_(kg)']

# Healthy_Diet_Lifestyle Interaction
df_new['HealthyDiet_Lifestyle'] = df_new['Healthy_Diet_Score'] * df_new['Lifestyle_Score']
df_new
```

We have a pair of columns with high coorelation to each other lets check them out.

The variables above are the pairs with the highest correlation to each other so the next step will be dropping one of the columns in each pair because we cannot have variables highly correlated to each other. 

```python

# Create a correlation matrix
correlation_matrix = df_new.corr().abs()

# Set the threshold for high correlation
threshold = 0.75

# Find pairs of highly correlated variables
high_corr_vars = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if correlation_matrix.iloc[i, j] >= threshold:
            high_corr_vars.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

# Drop a column from each pair of highly correlated variables
dropped_columns = []
for var1, var2 in high_corr_vars:
    if var1 in df_new.columns and var2 in df_new.columns:
        if correlation_matrix.loc[var1, var1] > correlation_matrix.loc[var2, var2]:
            dropped_columns.append(var2)
            df_new = df_new.drop(var2, axis=1)
        else:
            dropped_columns.append(var1)
            df_new = df_new.drop(var1, axis=1)

# Print the dropped columns
print("Dropped columns:", dropped_columns)
```

The columns "dropped columns" are the columns that were in pair with other variables and had high correlation to each other. The dropped columns are the columns we decided to drop from the pairs of highly correlated variables/columns.

The next step is ensembling, which is the use of an algorithm that makes use of more than one model to make a prediction. Random forest is an ensemble of decision trees that uses the ensemble algorithm in more than one decision tree to make a prediction. 

We have a total of five models counting the baseline model. Since the baseline model has the highest AUC score yet it had overfitting brings quite the dilemma so before building a random forest we will use other metrics apart from classification score to evaluate the model to establish which model is performing the best.

```python
# We used precision score,recall score, accuracy score and f1 score to determine which model is performing the best
def calculate_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    return precision, accuracy, recall, f1


predicted_labels_list = [test_pred,y_pred_2,y_pred_4,y_pred_6,y_pred_8]
# Assuming you have a list of predicted labels for each model

# Calculate the metrics
for i, predicted_labels in enumerate(predicted_labels_list):

    precision, accuracy, recall, f1 = calculate_metrics(y_test, predicted_labels)

    
    # Print the metrics for the current model
    print("Metrics for Model", i+1)
    print("Precision:", precision)
    print("Accuracy:", accuracy)
    print("Recall:", recall)
    print("F1-score:", f1)
    print()
```

Our baseline model has the metric results which is good but still not viable because we noticed the presence of overfitting as the model did not generalize on test data, so after some tree pruning we can see that our fifth model is the model with the best accuracy and so that is the model we will use with the Random Forest Classifier which is an ensemble of decision trees.
